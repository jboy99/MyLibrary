```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

In the cells above, we can see that the data contains 18 columns and 1000 rows and we can see excepts of the data entries.
Next we will perform univariate analysis(more like a frequency distribution of the itemns in each column) on each column,accompanied by diagramatic representations. 

```python
data_1 = data.loc[:,'What is your current age?'].value_counts()
print (data_1)
g = sns.countplot(x='What is your current age?', data=data)
```

```python
data_2 = data.loc[:,'Categories. SIMPLE'].value_counts()
print (data_2)
g = sns.countplot(x='Categories. SIMPLE', data=data)
```

```python
data_3 = data.loc[:,'How do you identify yourself?'].value_counts()
print (data_3)
g = sns.countplot(x='How do you identify yourself?', data=data)
```

```python
data_4 = data.loc[:,'Do you own or carry any forms of protection in case of assault or harm? SIMPLE'].value_counts()
print (data_4)
g = sns.countplot(x='Do you own or carry any forms of protection in case of assault or harm? SIMPLE', data=data)
```

```python
data_5 = data.loc[:,'Would you buy a product that would help protect YOU against assault? SIMPLE'].value_counts()
print (data_5)
g = sns.countplot(x='Would you buy a product that would help protect YOU against assault? SIMPLE', data=data)
```

```python
data_6 = data.loc[:,'Have you ever been a victim of sexual assault or rape?'].value_counts()
print (data_6)
g = sns.countplot(x='Have you ever been a victim of sexual assault or rape?', data=data)
```

```python
data_7 = data.loc[:,'Have you ever witnessed assault or rape?'].value_counts()
print (data_7)
g = sns.countplot(x='Have you ever witnessed assault or rape?', data=data)
```

```python
data_8 = data.loc[:,'If you answerd "Yes" or "Maybe" to any of the two previous columns, did you report to the hospital or police? SIMPLE'].value_counts()
print (data_8)
g = sns.countplot(x='If you answerd "Yes" or "Maybe" to any of the two previous columns, did you report to the hospital or police? SIMPLE', data=data)
```

```python
data_9 = data.loc[:,'How comfortable do you feel going to the authorities (police or hospitals) on a scale of 1 to 10 (where 1 = Very Uncomfortable to 10=Very Comfortable).'].value_counts()
print (data_9)
g = sns.countplot(x='How comfortable do you feel going to the authorities (police or hospitals) on a scale of 1 to 10 (where 1 = Very Uncomfortable to 10=Very Comfortable).', data=data)
```

The above univariate analyses definitely give us a sense of direction on the data, with regards to the responses to questions and the popularity of each response. Now we delve into bivirate analysis, to see the interplay between any two chosen parameters

```python
g = sns.countplot(x='How do you identify yourself?', hue = 'Have you ever been a victim of sexual assault or rape?', data=data)
```

```python
g = sns.countplot(x='How comfortable do you feel going to the authorities (police or hospitals) on a scale of 1 to 10 (where 1 = Very Uncomfortable to 10=Very Comfortable).', hue = 'Have you ever been a victim of sexual assault or rape?', data=data)
```
